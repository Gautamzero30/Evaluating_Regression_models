```python
import torch

X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)

y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1,1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1,1)
```

```python
from torch.utils.data import TensorDataset, DataLoader

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)
```

```python
import torch.nn as nn

class ANN(nn.Module):

    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(X_train.shape[1], 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, 16),
            nn.ReLU(),

            nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.network(x)
```

```python
model = ANN()
```

```python
criterion = nn.MSELoss()
```

```python
import torch.optim as optim

optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)
```

```python
epochs = 100

for epoch in range(epochs):

    model.train()

    for X_batch, y_batch in train_loader:

        optimizer.zero_grad()

        output = model(X_batch)

        loss = criterion(output, y_batch)

        loss.backward()

        optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1} | Loss = {loss.item():.4f}")
```

```python
model.eval()

with torch.no_grad():

    y_pred = model(X_test_tensor)
```

```python
y_pred_ANN = y_pred.numpy()
```

```python
print("MAE :", mean_absolute_error(y_test, y_pred_ANN))
print("MSE :", mean_squared_error(y_test, y_pred_ANN))
print("RMSE:", root_mean_squared_error(y_test, y_pred_ANN))
print("R² :", r2_score(y_test, y_pred_ANN))
```
